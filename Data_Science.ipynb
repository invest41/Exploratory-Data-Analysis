```
#Machine Learning #3:39am 12-02-2021
#1.0
import numpy as np, pandas as pd, sklearn.naive_bayes as sknb
import seaborn as sns, matplotlib.pyplot as plt
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier as dtc
from sklearn.metrics import accuracy_score as acsc
from sklearn.model_selection import train_test_split as tts
 
#2.0
X = np.array([
              [-1,1],
              [-2,-1],
              [-3,-2],
              [1,1],
              [2,1],
              [3,2]
])
 
Y = np.array([1,1,1,2,2,2])
 
X1,X2,Y1,Y2 = tts(X,Y,test_size=0.6)
print(f'Test Data: \n{X2}\n')
 
#3.0 Gausian - Naive Bayes
model1 = sknb.GaussianNB()
model1.fit(X1,Y1)
model1
prediction = model1.predict(X2)
score1 = (acsc(Y2,prediction))*100
print(f'Gausian prediction: {prediction}, {score1}% accuracy')
 
#4.0 Decision Tree Classifier
model2 = dtc()
model2.fit(X1,Y1)
model2
predict = model2.predict(X2)
score2 = (acsc(Y2,predict))*100
print(f'Decision Tree Prediction: {predict}, {score2}% accuracy')
 
 
#df = pd.DataFrame(data=X1)
#sns.scatterplot(x=df[0],y=df[1],)
```

```
#Machine Learning 5:42am, 13-02-2021
 
#1.0
import numpy as np, pandas as pd, sklearn.naive_bayes as sknb
import seaborn as sns, matplotlib.pyplot as plt
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier as dtc
from sklearn.metrics import accuracy_score as acsc
from sklearn.model_selection import train_test_split as tts
 
#2.0
df = pd.read_csv("/content/sample_data/california_housing_train.csv")
#df
 
#3.0
X = df.drop(columns=['median_income'])
Y = df[['median_income']]
#print(X,Y)
 
 
#4.0
X_train,X_test,Y_train,Y_test = tts(X,Y,test_size=0.6)
print(f'Test Data: \n{X_test}\n')
 
 
 
#5.0 Gausian - Naive Bayes
model1 = sknb.GaussianNB()
model1.fit(X_train,Y_train['median_income'])
model1
prediction = model1.predict(X_test)
score1 = (acsc(Y_test,prediction))*100
print(f'Gausian prediction: {prediction}, {score1}% accuracy')
 
 
 
 
 
 
#6.0 Decision Tree Classifier
model2 = dtc()
model2.fit(X_train, Y_train)
model2
predict = model2.predict(X_test)
score2 = (acsc(Y_test,predict))*100
print(f'Decision Tree Prediction: {predict}, {score2}% accuracy')
```
